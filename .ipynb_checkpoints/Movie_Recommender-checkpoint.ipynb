```python
#using the convert function to change the genre into a list of words.
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Function to convert the genre data into a list format
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj) :
        L.append(i['name'])
    return L
```

```python
#Applying the convert function to the genre column
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(5)
```

```python
#Applying the convert function to the keyword column
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(5)
```

```python
#Another function for converting the data in the cast column into a list format
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj) :
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
#Applying the convert3 funciton to the cast column.
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(5)
```

```python
#Definning a function to convert crew data into a list format for every row.
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj) :
        if i['job'] == 'Director' :
            L.append(i['name'])
            break
    return L
```

```python
#Applying the FetchDirector function for the crew column.
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#using the count vectoriser module from sklearn to make a prediction model.
from sklearn.feature_extraction.text import CountVectorizer
#Eliminating any stop words ex:on, the , at, or,....
#Setting max word count as 5000 for every movie
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
#similar words "stemming elimination"
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
#for higher dimensional data euclidian distance measure fails so we are using the
#cosine distance measure method
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Calculated distances between every movie using the cosine distance module from the sklearn library
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
#converting the index we recieved from the similarity score list into the respective movie name from the dataframe
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
              
              
    return
```
